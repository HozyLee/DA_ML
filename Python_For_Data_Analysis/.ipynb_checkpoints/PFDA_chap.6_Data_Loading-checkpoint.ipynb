## 6.1.5 XML 과 HTML : 웹 내용 긁어오기(크롤링)
### lxml : 큰 파일 빠르게 처리, 인터페이스 제공
### lxml.html : HTML 처리
### lxml.objectify : XML 처리

### 순서 : 데이터 url 확인 -> requests 모듈로 불러오기 -> lxml 로 파싱

### 파싱(parsing) 이란? 데이터 타입을 알아서 구별해 주는 기능
#### 웹 페이지에서 긁어오면 모두 문자열로 취급되기 때문에 파싱이 필요하다.

```python
import requests

from lxml.html import parse
from io import StringIO

text = requests.get('http://finance.yahoo.com/q/op?s=AAPL+Options').text # requests를 통해 해당 url의 데이터를 text 형식으로 불러온다.
head = requests.get('http://finance.yahoo.com/q/op?s=AAPL+Options').headers # requests를 통해 해당 url의 헤더를 text 형식으로 불러온다.

parsed = parse(StringIO(text)) # 저장한 객체를 StringIO를 통해 변환한다음 파싱하여 객체에 저장한다. 

doc = parsed.getroot() # 파싱한 객체를 getroot()를 통해 저장한다. doc 객체에는 모든 HTML 태그가 추출되어 있음
```

### th : 헤더가 들어있다
### td : 데이터가 들어있다

```python
def _unpack(row , kind = 'td'):
    elts = row.findall('.//%s' % kind)
    return [val.text_content().strip().split('\n')[0] for val in elts]
```

### 위의 함수를 사용하여 데이터를 추출할 수 있다.

```python
_unpack(rows[0], kind = 'th')
```

```python
_unpack(rows[2], kind = 'td')
```

### TextParser 클래스 : pandas 에서 제공하는 자동형 변환 클래스

```python
from pandas.io.parsers import TextParser
```

```python
def parse_options_data(table):
    rows = table.findall('.//tr')
    header = _unpack(rows[0], kind = 'th') # _unpack 함수를 사용하여 첫번째 행(rows[0])을 header로 추출
    data = [_unpack(r) for r in rows[1:]] # _unpack 함수를 사용하여 두번째 행(rows[:1])을 data로 추출
    return TextParser(data, names = header).get_chunk() # TextParser 를 사용하여 데이터를 넣고, header 를 지정하여 데이터 타입을 알아서 추측
```

```python
call_data = parse_options_data(calls)
```

```python
put_data = parse_options_data(puts)
```
