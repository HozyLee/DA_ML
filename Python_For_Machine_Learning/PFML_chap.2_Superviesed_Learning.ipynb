### KNeighborsClassfier 분석

```python
fig, axes = plt.subplots(1,3, figsize = (10,3))

for k, ax in zip([1,3,9], axes):
    
    clf = KNeighborsClassifier(n_neighbors = k).fit(X_f_train, y_f_train)
    mglearn.plots.plot_2d_separator(clf, X_f_train, fill = True, eps = 0.5, ax = ax, alpha = .4) # 모델에서 구분한 클래스 구분 선 그리기
    mglearn.discrete_scatter(X_f_train[:,0], X_f_train[:,1], y_f_train, ax = ax) # 겹쳐서 그리기
    
    ax.set_title('KNN ( k = {} )'.format(k))
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
    
axes[0].legend(loc = 3)
```

### KNeighborsRegressor 분석

```python
fig, axes = plt.subplots(1,3, figsize = (15,4))
# -3 과 3 사이에 1000개의 데이터포인트를 만든다.

line = np.linspace(-3,3,1000).reshape(-1,1)
for n_neighbors, ax in zip([1,3,9], axes):
    # 1,3,9 개의 이웃을 사용하여 예측
    reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors = n_neighbors)
    reg.fit(X_w_train, y_w_train)
    ax.plot(line, reg.predict(line)) # 위의 데이터 포인트를 X 값으로 주었을 때 예측값에 대한 그래프
    ax.plot(X_w_train, y_w_train, '^' , c = mglearn.cm2(0), markersize = 8)
    ax.plot(X_w_test, y_w_test, 'v', c = mglearn.cm2(1), markersize = 8)
    
    ax.set_title(
        "{} neighbors train score : {:.2f} test score : {:.2f}".format(
        n_neighbors, reg.score(X_w_train, y_w_train), reg.score(X_w_test, y_w_test)))
    ax.set_xlabel('feature')
    ax.set_ylabel('target')
    axes[0].legend(['model prediction','train data/ target ','test data/ target'],loc = 'best')
```

### 릿지 회귀(Ridge regression) 
#### sklearn.linear_model.Ridge( alpha =   , ....)
####  과적합이 되지 않도록 각 feature 의 coefficient 를 구할 때 , '제약조건'(regularization) 을 추가한다.
#### 결과적으로 중요한 특성값이 아닌 값들의 coefficient 는 대부분 0에 가까워진다.
#### 옵션
##### alpha : [0,1] 범위의 수로 클수록 coefficient의 절댓값 크기가 작아진다(일반화, smooth 한 그래프 )

```python
from sklearn.linear_model import Ridge

rid = Ridge().fit(X_b_train, y_b_train)

# 모델 평가
print('ridge 회귀에 대한 boston 데이터 평가')
print('train 데이터 스코어 : {:.2f}'.format(rid.score(X_b_train, y_b_train)))
print('test 데이터 스코어 : {:.2f}'.format(rid.score(X_b_test, y_b_test)))

# 위의 기본 linear regression 보다 과적합이 덜 일어나며, test 데이터에 대해서도 예측률이 올라간 것을 알 수 있다.
```

```python
# alpha 값에 따른 예측률
rid01 = Ridge(alpha = 0.1)
rid10 = Ridge(alpha = 1.0)
```

```python
rid01.fit(X_b_train, y_b_train)
rid10.fit(X_b_train, y_b_train)
```

```python
print('Ridge 회귀 ( alpha = 0.1 ) 평가')
print('train set score : %.2f' % (rid01.score(X_b_train, y_b_train)))
print('test set score : %.2f' % (rid01.score(X_b_test, y_b_test)))
```

```python
print('Ridge 회귀 ( alpha = 1.0 ) 평가')
print('train set score : %.2f' % (rid10.score(X_b_train, y_b_train)))
print('test set score : %.2f' % (rid10.score(X_b_test, y_b_test)))

# 둘다 안 좋아짐 : train set 에서도, test set 에서도
```

### 라소 회귀(Lasso regression)
### sklearn.linear_model.Lasso( alpha =  ,.... )
#### L1 규제 : 중요하지 않은 특성의 coefficient 를 0으로 만든다.
#### 특성선택이 자동으로 이루어진다.
#### 모델에서 중요한 특징값이 무엇인지 파악하기 쉽다.

### 라소 vs 릿지
#### 라소 : coefficient 를 0으로 만든다.
#### 릿지 : coefficient 를 0에 근사시킨다.

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_b_train, y_b_train)
```

```python
print('Lasso 회귀를 이용한 평가')
print('train set score : {:.2f}'.format(lasso.score(X_b_train,y_b_train)))
print('test set score : {:.2f}'.format(lasso.score(X_b_test,y_b_test)))
print('사용한 feature 의 수 :{} 개'.format(np.sum(lasso.coef_ != 0 ))) # 회귀 결과에서 coefficient 가 0이 아닌것의 갯수
## 겁나 별론데;;;
```

```python
# 릿지 회귀와 마찬가지로 alpha 옵션을 통해 coeffient 의 수준을 설정할 수 있다.
# max_iter 를 증가시키지 않으면 늘리라는 경고가 발생한다.
lasso001 = Lasso(alpha = 0.01 , max_iter = 100000).fit(X_b_train, y_b_train)
```

```python
print('Lasso001 회귀를 이용한 평가')
print('train set score : {:.2f}'.format(lasso001.score(X_b_train,y_b_train)))
print('test set score : {:.2f}'.format(lasso001.score(X_b_test,y_b_test)))
print('사용한 feature 의 수 :{} 개'.format(np.sum(lasso001.coef_ != 0 )))

# alpha(제약) 을 줄여줌으로써 사용하는 특성 수가 증가하고 그에 따라 일반화된 모델이 되어 예측값이 증가한 것을 알 수 있다.
```

```python
# 하지만 alpha 를 너무 낮추면 모든 특성값을 다 이용하므로 과대적합이 되므로 LinearRegrassion 의 결과와 비슷해진다.
lasso00001 = Lasso(alpha = 0.001, max_iter = 100000).fit(X_b_train, y_b_train)
print('Lasso00001 회귀를 이용한 평가')
print('train set score : {:.2f}'.format(lasso00001.score(X_b_train,y_b_train)))
print('test set score : {:.2f}'.format(lasso00001.score(X_b_test,y_b_test)))
print('사용한 feature 의 수 :{} 개'.format(np.sum(lasso00001.coef_ != 0 )))

# test 데이터에 대해 예측력이 떨어진다. 즉, 과적합으로 인해 일반화를 잘 못하는 경우이다. 
```

```python
plt.plot(lasso.coef_ , 's' , label = 'Lasso alpha = 1')
plt.plot(lasso001.coef_ , '^' , label = 'Lasso alpha = 0.01')
plt.plot(lasso00001.coef_ , 'v' , label = 'Lasso alpha = 0.0001')

plt.plot(rid01.coef_, 'o' , label = 'Ridge alpha = 0.1')

plt.legend(ncol = 2, loc = (0,1.05))
plt.ylim(-25,25)
plt.xlabel('coeifficient')
plt.ylabel('value of coefficient')

# 제약(alpha 가) 큰 모델일수록 coefficient 의 절댓값 크기가 작아진다.
```

### 분류용 선형모델
### 이진분류
#### 계산한 초평면을 기준으로 각 데이터의 부호에 따라 두개의 클래스로 분류
#### 1. 로지스틱 회귀 (LogisticRegression)
#### 2. 서포트벡터머신(svm.LinearSVC)

#### 매개변수 C : 위의 lasso 와 ridge 와 비슷한 제약 매개변수 / 작동은 반대로
##### c 증가 : train 데이터에 맞추려고 함 / 개개의 데이터 포인트를 고려 / 너무 높은 경우 과적합
##### c 감소 : w( 계수 벡터 ) 를 0 에 맞추려고 함 / 대략적인 분류 기준 고려


### 위의 회귀 모형과 매개변수 비교

```python
# 적합성 증가를 위해 c 의 값을 증가시켜보겠다.(증가 -> 일반화 감소)
logreg100 = LogisticRegression(C = 100).fit(X_c_train, y_c_train)

print('LogisticRegression c = 100 평가')
print('train set score : %.3f' % (logreg100.score(X_c_train, y_c_train)))
print('test set score : %.3f' % (logreg100.score(X_c_test, y_c_test)))
# 이 데이터 셋에서는 조금 더 복잡한(여러 특성을 고려한) 모델이 더 적합하다는 것을 보여준다.
```

```python
logreg001 = LogisticRegression(C = 0.01).fit(X_c_train, y_c_train)

print('LogisticRegression c = 0.01 평가')
print('train set score : %.3f' % (logreg001.score(X_c_train, y_c_train)))
print('test set score : %.3f' % (logreg001.score(X_c_test, y_c_test)))
```

```python
# C 설정에 따른 계수의 크기
plt.plot(logreg.coef_.T , 'o', label = "C = 1")
plt.plot(logreg100.coef_.T , '^', label = "C = 100")
plt.plot(logreg001.coef_.T , 'v', label = "C = 0.01")

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
plt.hlines(0,0, cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel('feature')
plt.ylabel('abs coefficient of feature')
plt.legend()

# c 가 크면 각 특성값을 고려하므로 coeifficient 의 크기가 큰 것을 알 수 있다.
# logistic regression 은 L2 규제를 사용한다.
# 계수들을 0과 근사시키지만 0이 되지는 않는다.
# c 의 크기에 따른 각 feature 들의 형태를 보면, feature 가 target 에 어떤 영향을 주는지 알 수 있다.
```

```python
# L1 규제를 사용할 때 정확도
for C, marker in zip([0.001,1,100], ['o','^','v']):
    lr_l1 = LogisticRegression(C = C).fit(X_c_train, y_c_train)
    print('C = {:.3f} 인 L1 규제의 로지스틱 회귀의 train set에서의 정확도 : {:.3f}'.format(C , lr_l1.score(X_c_train, y_c_train)))
    print('C = {:.3f} 인 L1 규제의 로지스틱 회귀의 test set에서의 정확도 : {:.3f}'.format(C , lr_l1.score(X_c_test, y_c_test)))
    print('='*80)
    plt.plot(lr_l1.coef_.T, marker, label = 'C = {:.3f}'.format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names , rotation = 90)
plt.hlines(0,0, cancer.data.shape[1])
plt.xlabel('feature')
plt.xlabel('abs coefficient of feature')

plt.ylim(-5,5)
plt.legend()

# C 가 증가할수록 각 coefficient 가 살아남아(0에 근접하지 않아) train set 에 대해 더 과적합되는 것을 볼 수 있다.
```

## 2.3.5 결정 트리
#### CART(Classification And Regression Tree) : 분류와 회귀에 둘 다 사용
#### 각 노드의 질문에 y/n 의 대답으로 edge 를 이어나가 마지막에 leaf 를 만든다.

### 결정 트리 만들기
- 결정 트리 학습 : 정답에 가장 빨리 도달하는 y/n 질문 목록(test)을 만든다는 것
- 연속 데이터인 경우 질문 목록 : 각 데이터 포인트는 S_i( i feature 의 기준 값 ) 보다 큰가? y/n
    - 즉, 각 test 는 하나의 feature 로 데이터를 둘로 나누는 것
    - 순수 노드 : target 하나로만 이루어진 node 

### 결정 트리 복잡도 제어
- 모든 leaf node = pure node : 과적합
    - 1. 사전 가지치기 : 최대 깊이 제한, node 에서 최소한 가지고 있어야 할 데이터 포인트 갯수 설정
    - 2. 사후 가지치기 : pure node 까지 간 후 과적합인 node 를 삭제하거나 병합
    
    scikit-learn 은 1 번만 지원


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

cancer = sklearn.datasets.load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target , random_state = 42)

tree = DecisionTreeClassifier(random_state = 0)
tree.fit(X_train, y_train)

print('accuracy for train dataset : {:.2f}'.format(tree.score(X_train,y_train)))
print('accuracy for test dataset : {:.2f}'.format(tree.score(X_test,y_test)))

# train set 에 대해 정확도가 1이 나온것으로 보아 leaf 가 pure node 라는 것을 예상할 수 있다. 과적합
```

```python
# max_depth 를 설정하여 연속한 질문의 갯수를 제한한다. 트리의 최대 높이를 제한하여 과적합을 막는다.
tree = DecisionTreeClassifier(max_depth = 4)
tree.fit(X_train,y_train)

print('accuracy for train dataset : {:.2f}'.format(tree.score(X_train,y_train)))
print('accuracy for test dataset : {:.2f}'.format(tree.score(X_test,y_test)))
# test set 에 대한 정확도가 증가하였다. 즉, 과적합이 조금 방지되고, 일반화가 진행되었다. 
```

## 2.3.6 결정 트리의 앙상블
### 앙상블 : 여러 모델을 연결하여 강력한 모델을 만드는 기법
#### 모델의 약점을 보완
### 1. 랜덤 포레스트
### 2. 그레디언트 부스팅  : 약한 분류기들의 선형 가중치의 결합(가중치의 합 = 1)

### 랜덤 포레스트
#### 서로 다른 성격의 결정 트리의 묶음 / 서로 구별되어야 한다. ( 무작위성 )
#### 무작위성 주입
##### 1. 데이터 포인트 랜덤
##### 2. 특성 무작위 선택

### 랜덤 포레스트 구축
#### 생성할 트리의 갯수( n_estimator )
#### 무작위성(부스트트랩 샘플) : 무작위로 샘플 수 만큼 반복 추출
#### 이 데이터 셋으로 각각의 결정 트리 생성
#### 차이점 : 트리 생성 시 좋은 특성으로 만드는 것이 아니라, 특성을 먼저 랜덤하게 추출하고 그것에서 최선의 node 를 짠다. ( max_feature 로 특성 수 정할 수 있다.)
#### 위 과정은 노드마다 수행된다.
#### 즉, max_feature 가 크다면, 여러 feature 중 고려할 feature 의 수가 많은 것이므로 트리들끼리 비슷한 feature 를 사용하여 유사해지고
#### max_feature 가 작다면, tree끼리 feature 가 겹칠 확률이 적어지므로 성격이 달라진다. 각 트리의 깊이가 깊어진다.

### 랜덤 포레스트 분석


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)
X_train, X_test, y_train, y_test = train_test_split(X , y , stratify = y, random_state = 42)

rdforest = RandomForestClassifier(n_estimators =5, random_state = 2) # 개별 결정나무의 수 = 5
rdforest.fit(X_train, y_train)
```

```python
# 개별 결정나무의 분류법을 그리고
# 그것을 종합한 randomForest 의 분류법을 그린다.

fig, axes = plt.subplots(2,3, figsize = (20,10))

for i, (ax,tree) in enumerate(zip(axes.ravel(), rdforest.estimators_)): # axes.ravel() : fig 안의 각 axes 의 객체 리스트를 반환
    ax.set_title('Tree {}'.format(i))
    mglearn.plots.plot_tree_partition(X, y, tree , ax = ax)
    
mglearn.plots.plot_2d_separator(rdforest, X, fill = True, ax = axes[-1,-1], alpha = .4)
axes[-1,-1].set_title('Random Forest') # fig 안의 마지막 axes : axes[-1,-1]
mglearn.discrete_scatter(X[:,0] , X[:,1], y)
```

```python
# 유방암 데이터 셋에 대한 random forest 성능
X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(cancer.data, cancer.target , random_state = 0)
rdforest_c = RandomForestClassifier(n_estimators = 100, random_state = 0)
rdforest_c.fit(X_c_train, y_c_train)

print('Random forest 성능 측정(유방암 데이터 셋)')
print('accuracy for train dataset : {:.2f}'.format(rdforest_c.score(X_c_train, y_c_train)))
print('accuracy for test dataset : {:.2f}'.format(rdforest_c.score(X_c_test, y_c_test)))

# max_feature 처럼 중요한 설정을 하지 않아도 기본적으로 단일 트리보다 좋은 성능이 나온다는 것을 알 수 있다.
```

## 2.3.7 커널 서포트 벡터 머신

```python
# 특성 1의 제곱 데이터를 특성으로 추가하여, 특성을 확장한다. (X_new 데이터)
## 특성값들 (특성0, 특성1, 특성1 **2 ) 으로 모델을 학습시켜본다.

X_new = np.hstack([X, X[: , 1:] **2 ])
```

```python
from mpl_toolkits.mplot3d import Axes3D, axes3d

figure = plt.figure()
# 3차원 그래프

ax = Axes3D(figure, elev = -152, azim = -26)
# y ==0 인 포인트를 먼저 그리고 그다음 y == 1 인 포인트를 그린다.


mask = ( y == 0 )
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c = 'b', 
          cmap = mglearn.cm2, s = 60, edgecolor = 'k') 
# mask 를 씌워(값이 0인 것들)로 하여 각 feature(3개)를 각 축의 값으로 준다.
# mask 에서 True 에 해당하는 값 (짝수)
# 파란 동그라미 표시

ax.scatter(X_new[~mask, 0],X_new[~mask, 1],X_new[~mask, 2], c = 'r', marker = '^',
           cmap = mglearn.cm2, s = 60, edgecolor = 'k') 
# mask 에 반대되는 것들 
# mask 에서 False 에 해당하는 값(홀수)
# 빨간 세모 표시

ax.set_xlabel('feature 0')
ax.set_ylabel('feature 1')
ax.set_zlabel('feature 1 squared')

# 위의 2차원 그래프와 비교해보면, x,y 평면쪽으로 정사영시킨 결과와 같다.
```

```python
# 이제 3차원 공간의 평면을 사용하여 두 클래스를 구분할 수 있다.
linear_svm_3d = LinearSVC().fit(X_new, y)
coef , intercept = linear_svm_3d.coef_.ravel() , linear_svm_3d.intercept_
```

```python
###### 선형 결정 경계 그리기
figure = plt.figure()
ax = Axes3D(figure, elev = -152, azim = -26)
xx = np.linspace(X_new[:,0].min() - 2, X_new[:,0].max() + 2, 50)
yy = np.linspace(X_new[:,1].min() - 2, X_new[:,1].max() + 2, 50)

XX, YY = np.meshgrid(xx,yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX,YY,ZZ, rstride = 8, cstride = 8, alpha = .3) # 결정경계를 그리는 메서드

mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c = 'b', 
          cmap = mglearn.cm2, s = 60, edgecolor = 'k') # mask 를 씌워(값이 0인 것들)로 하여 각 feature(3개)를 각 축의 값으로 준다.

ax.scatter(X_new[~mask, 0],X_new[~mask, 1],X_new[~mask, 2], c = 'r', marker = '^',
           cmap = mglearn.cm2, s = 60, edgecolor = 'k')

ax.set_xlabel('feature 0')
ax.set_ylabel('feature 1')
ax.set_zlabel('feature 1 squared')
```

```python
# 원래 있는 2d feature(feature0, feature1) 로 투형해보면 더 이상 결정경계가 선형이 아니다. 
# 타원에 가까운 모습을 보여준다.
ZZ = YY ** 2

dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels = [dec.min(), 0, dec.max()],
            cmap = mglearn.cm2, alpha = .5)

mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel('feature 0')
plt.ylabel('feature 1')
```

### SVM 매개변수 튜닝
#### 1. gamma : 가우시안 커널 폭의 역수 / 하나의 훈련 새믈이 미치는 영향의 범위 조정 / 커질수록 번위 제한
#### 2. C : 규제 매개변수 / 각 포인트의 중요도를 제한한다.

```python
# 매개변수에 따른 변화
fig, axes = plt.subplots(3,3, figsize = (15,10))

for ax, C in zip(axes , [-1,0,3]):
    for a, gamma in zip(ax, range(-1,2)):
        mglearn.plots.plot_svm(log_C = C, log_gamma = gamma , ax = a)
        
        
axes[0,0].legend(['class 0','class 1',' support vector of class 0','support vector of class 1'], ncol = 4, loc = (.9,1.2))

# gamma 증가 (왼 -> 오) : 각 데이터 포인트에 민감(부드러운 곡선에서 -> 복잡한 곡선으로)
# C 증가 (위 -> 아래) : 제약이 큰 모델 , 각 데이터 포인트의 영향력이 작아진다.( 더 복잡해진다. )
```

### SVM 을 위한 데이터 전처리
#### 특성값의 범위가 비슷해지도록 조정하는 방법(0 과 1 사이)
#### MinMaxScaler( ) 메서드 사용

```python
# train data 에서 특성별 최솟값 계산
min_on_training = X_train.min(axis = 0)

# train data 에서 특성별 범위 계산(max - min)
range_on_training = (X_train - min_on_training).max(axis = 0)

# train data 에서 최솟값을 빼고 범위로 나누면
# 각  특성에 대해 최솟값은 0, 최댓값은 1이다.
X_train_scaled = (X_train - min_on_training) / range_on_training
print('특성별 최솟값 \n {}'.format(X_train_scaled.min(axis = 0)))
print('특성별 최댓값 \n {}'.format(X_train_scaled.max(axis = 0)))
```

```python
#  test data에도 같은 작업을 적용하지만 
# train data 에서 계산한 최솟값과 범위를 사용한다.
X_test_scaled = (X_test - min_on_training) / range_on_training

```

```python
# scale 처리가 된 train input 으로 모델을 fitting 하고 
# test input 으로 과적합을 평가한다.
svc = sklearn.svm.SVC()
svc.fit(X_train_scaled , y_train)

print("Accuracy on scaled train set : {:.3f}".format(svc.score(X_train_scaled, y_train)))
print("Accuracy on scaled test set : {:.3f}".format(svc.score(X_test_scaled , y_test)))

# scale 처리한 데이터셋으로 모델을 fitting 하니 좋은 결과가 나왔다.
```
