## 2.3.8 신경망(딥러닝) p.138~
### 다층 퍼셉트론(피드 포워드 신경망) : 간단한 분류,회귀에 사용할 수 있다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
import sklearn
from sklearn.model_selection import train_test_split
from mglearn.plots import plot_2d_separator
from mglearn import discrete_scatter
```

## 신경망 모델
### 선형 모델의 일반화 : 여러 선형 모델의 단계를 거쳐 의사결정
### 선형 모델 : 각각 특성값과 가중치의 선형 합

### 다층 퍼셉트론(MLP) : 가중치의 합을 만드는 과정이 여러번 반복
#### 여러 개의 가중치 합을 계산하는 것 = 하나의 가중치 합을 계산하는 것
#### 선형 모델과 유사하여, 더 강력하게 만들기 위해서 기교가 필요
#### 가중치 합 계산 후 비선형 함수를 적용 / 함수(활성화 함수)에 적용되어 나온 값이 노드의 output 이 된다.
#### 1. 렐루 합수 : 0 이하는 잘라버린다. / 0보다 작은 값은 0으로 , 큰 값은 그대로
#### 2. 하이퍼볼릭 탄젠트 함수 : 낮은 값은 -1로 수렴, 큰 값은 +1 로 수렴
#### 3. 시그모이드 함수 : 1 / (1 + exp(-z)) ( z 가중치와 입력변수의 선형 합 : 즉 선형분리함수의 output)
#### 이를 통해 선형함수보다 더 복잡한 모델을 만들 수 있다.

```python
line = np.linspace(-3,3,100)
plt.plot(line, np.tanh(line) , label = 'tanh') # hyper tangent 를 적용하여 그래프를 그린다.
plt.plot(line, np.maximum(line, 0), label = 'relu') # 0을 최솟값으로 정의하는 함수 np.maximum 을 적용하여 렐루 함수 그래프를 그린다.
plt.legend(loc = 'best')
plt.xlabel('x')
plt.ylabel('relu(x), tanh(x)')
```

### 중요한 매개변수 : 은닉층의 유닛의 개수
#### 데이터의 규모가 커질수록 유닛의 수가 많아진다(복잡도 증가)
#### 또는 은닉층의 수를 추가할 수도 있다. ( 수많은 은닉층의 신경망을 딥러닝 모델이라고 부른다.)

### 신경망 튜닝
#### MLPClassifier : MLP(다층 퍼셉트론) 구현
#### 옵션 값들
##### hidden_layer_sizes : 은닉층 당 은닉노드의 수 ( default : [100,] / 은닉층 1개, 은닉노드 100개)
##### activation : 사용할 활성화 함수( default : 'relu' )
##### solver : 최적화 방법( default : 'adam' )
##### learning_rate : 학습률 / learning_rate_init : 초기 학습률 값
##### alpha : L2 규제를 통해 가중치를 0에 가깝게 근사시켜 모델의 복잡도를 줄여준다.
##### random_state : 가중치의 초깃값을 정해준다. (초깃값에 따라 모델의 학습이 바뀐다)
##### max_iter : 최대 가능한 iteration 수 (default : 200번)

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

```python
make_moons()
# 두개의 feature 에 output 이 0,1 인 이진분류 데이터
```

```python
# 데이터 셋을 불러와 선언한다.
X, y = make_moons()

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, stratify = y, random_state = 42)

# 다계층 퍼셉트론(Many Layered Perceptron) 을 만들고 train data 로 모델을 fitting 한다.
mlp = sklearn.neural_network.MLPClassifier(solver = 'lbfgs' , random_state = 0) #  최적화 함수 'lbfgs' 사용
mlp.fit(X_train, y_train)


mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train) # 두개의 특성을 각각 2차원으로 그리고 좌표를 찍은 후 y_train 에 맞게 0,1 로 표시
plt.xlabel('feature 0')
plt.ylabel('feature 1')

# 선형 함수의 연속임에도 불구하고 비교적 매끄러운 결정경계를 만들었다.
```

### 기본적으로 MLP의 default 은닉노드의 수는 100개이다.
#### 데이터 셋의 복잡도와, 크기에 따라 은닉노드의 수를 정하여 모델을 만들 수 있다.
#### 옵션 : hidden_layer_sizes = [ 1번 은닉층의 은닉노드 수 , 2번 은닉층의 은닉노드의 수 , .... ]

```python
mlp = sklearn.neural_network.MLPClassifier(solver = 'lbfgs', random_state = 0, 
                                                          hidden_layer_sizes = [10]) # 은닉층 1 개 은닉노드 수 10개로 setting 한다.

mlp.fit(X_train, y_train)

mglearn.plots.plot_2d_separator(mlp, X_train, fill = True , alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)

plt.xlabel('feature 0')
plt.ylabel('feature 1')

# 은닉 노드의 수가 줄어들어 결정경계가 덜 매끄러워졌다(날카롭)
```

```python
# 은닉 층: 2 / 은닉노드 수 : 10, 10 인 모델로 해보자
mlp = sklearn.neural_network.MLPClassifier(solver = 'lbfgs', random_state = 0, 
                                                          hidden_layer_sizes = [10, 10])

mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)

plt.xlabel('feature 0')
plt.ylabel('feature 1')
```

```python
# 은닉 층: 2 / 은닉노드 수 : 10, 10 인 모델 / + 활성화 함수 : tanh 사용
mlp = sklearn.neural_network.MLPClassifier(solver = 'lbfgs', random_state = 0, 
                                                          hidden_layer_sizes = [10, 10],
                                                          activation = 'tanh') # 활성화 함수로 tanh 사용 : 더 매끄러운 경계

mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)

plt.xlabel('feature 0')
plt.ylabel('feature 1')

# tanh (hyper tangent function ) 사용으로 결정경계가 더 매끄러워졌다.

# 특성의 수가 적은 데이터에서는 (여기서는 feature 수가 2(차원))
# 은닉층과 은닉노드의 수를 늘리고
# 유연한 활성화 함수를 사용 (tanh 같은)
# 하여 문제를 해결할 수 있다.
```

```python
# L2 규제를 사용(가중치를 0에 근사시키기) 하여 모델의 복잡도 줄이기
# alpha 옵션 사용

fig , axes = plt.subplots(2, 4, figsize = (20, 8))

for axx, n_hidden_nodes in zip(axes, [10,100]): # axx : axes[0][] , axes[1][] 
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]): # ax : axx[][0] ~ axx[][3] (axes[][0] ~ axes[][3]) / axx[]
        mlp = MLPClassifier(solver = 'lbfgs', random_state = 0,
                                   hidden_layer_sizes = [n_hidden_nodes, n_hidden_nodes], # 은닉 층 (10,10) 과 (100,100)
                                   alpha = alpha)
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3, ax = ax)
        mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train, ax = ax)
        ax.set_title('n_hidden = [{}, {}] \nalpha = {:.4f}'.format(n_hidden_nodes, n_hidden_nodes, alpha))
        
# alpha (MLPClassifier 의 옵션) 이 증가할 수록 영향이 적은 가중치는 0에 근사하여 모델이 더욱 단순해진다. (복잡도 감소)
# 즉, 결정 경계의 굴곡이 작아진다. 
# 결론 : alpha 증가 -> 결정경계 복잡도 감소(각각의 점보다 "전체적인 추세"에 따라 결정 경계 그림)

# 현재 feature 수가 2개 밖에 없고, 두 class 의 형태가 복잡(U자형) 이므로 
## alpha 를 너무 높이면 잘 분류하지 못한다.
```

### 신경망 모델의 복잡도 제어
#### 1. 은닉층 / 은닉 노드의 수 : hidden_layer_sizes = [ , , , ]
#### 2. alpha(규제) : 클수록 영향력이 적은 가중치 0에 근사 -> 복잡도 감소(전체적 추세 반영)

### 신경망 모델의 학습
#### 가중치의 무작위 선택
#### 가중치 초깃값 : 학습에 영향

```python
# 가중치 초깃값이 다른 모델들의 학습의 차이
# random_state ( 가중치의 초깃값 설정)

fig, axes = plt.subplots(2, 4, figsize = (20,8))
for i, ax in enumerate(axes.ravel()):
    mlp = sklearn.neural_network.MLPClassifier(solver = 'lbfgs', random_state = i,
                               hidden_layer_sizes = [100, 100]) 
    mlp.fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3, ax = ax)
    mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train, ax = ax )
    
    
# random_state : 가중치의 초깃값에 따라 학습(결정 경계 그리기)이 바뀐다.
```

```python
# 유방암 데이터를 신경망 모델에 적용해보자
# 신경망 모델은 input의 scale 차이에 민감하다(성능 좌우)
cancer = sklearn.datasets.load_breast_cancer()

print('유방암 데이터의 feature 별 최댓값 : \n {}'.format(cancer.data.max(axis = 0))) # 각 feature 별 최댓값들
```

```python
X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(cancer.data, cancer.target, random_state = 0)

mlp = MLPClassifier(random_state = 42)
mlp.fit(X_c_train, y_c_train)

print('Accuracy on train data: {:.2f}'.format(mlp.score(X_c_train, y_c_train)))
print('Accuracy on test data: {:.2f}'.format(mlp.score(X_c_test, y_c_test)))
```

### 신경망 모델의 입력값
#### SVC 와 마찬가지로 입력값의 범위와 scale 에 따라 성능 차이가 심하다.
#### 신경망 모델은 input 을 평균 0 , 분산 1 로 맞추어 주는 것이 좋다.(표준정규분포와 근사한 데이터에 성능이 좋다.)
#### 입력 변수의 타입
##### 범주형 : 각 범주의 빈도가 비슷하면 좋고 한쪽으로 치우치지 않는 것이 좋다.
##### 연속형 : 대칭 분포인 경우가 좋다. (대칭분포이며 mean = 0, std = 1 )
##### 비대칭인 경우 scale 조정이나 범주화를 통해 빈도를 조정하여 대칭성을 만들어준다.

```python
## Input Data scaling : 각 feature 를 평균 0 , 분산 1 로 바꾸어 scaling 해준다.
## ""train 데이터""에서 얻은 기준(mean, std) 로 scaling 을 해준다.

# train data 의 각 feature 의 평균을 계산한다.
mean_on_train = X_c_train.mean(axis = 0)
# train data 의 각 feature 의 표준 편차 계산
std_on_train = X_c_train.std(axis = 0)

# (X - mean) / std 로 정규화 시켜준다. ~N(0,1) : 표준정규분포
X_c_train_scaled = (X_c_train - mean_on_train) / std_on_train
# train data 의 평균, 표준편차를 이용해 test data 에도 적용시켜준다.
X_c_test_scaled = (X_c_test - mean_on_train) / std_on_train

mlp = MLPClassifier(random_state = 0)
mlp.fit(X_c_train_scaled, y_c_train)

print('Accuray on Scaled Train data : {:.3f}'.format(mlp.score(X_c_train_scaled, y_c_train)))
print('Accuray on Scaled Test data : {:.3f}'.format(mlp.score(X_c_test_scaled, y_c_test)))

# 조정 전의 데이터에서 보다 성능이 훨씬 좋아졌다.
## 경고 : iteration(반복) 최대 반복횟수에 도달하였다. solver: 'adam' 에서 나온것
### 해결법 : 반복횟수(max_iter) 를 늘려주어야 한다.
```

```python
# max_iter = 1000 으로 주어 iteration 수를 증가시켜준다.(default 값 : 200번)

mlp = MLPClassifier(random_state = 0, max_iter = 1000)
mlp.fit(X_c_train_scaled, y_c_train)

print('Accuray on Scaled Train data : {:.3f}'.format(mlp.score(X_c_train_scaled, y_c_train)))
print('Accuray on Scaled Test data : {:.3f}'.format(mlp.score(X_c_test_scaled, y_c_test)))

# 반복횟수 조정을 통해 성능을 더 좋게 만들었다.
```

## 2.4.1 결정함수

## 2.4.3 다중 분류의 불확실성
### 다중분류에서도 사용 가능
#### decision_function()
#### predict_proba : 각 데이터 포인트는 속할 확률(predict_proba)이 가장 큰 클래스로 분류된다.

```python
from sklearn.datasets import load_iris

iris = load_iris()
X_train , X_test , y_train , y_test = train_test_split(iris.data, iris.target, random_state = 42)

gbrt = GradientBoostingClassifier(learning_rate = .01, random_state = 0)
gbrt.fit(X_train, y_train)
```

```python
print('결정 함수의 결과 형태 : {}'.format(gbrt.decision_function(X_test).shape))
# (sample 의 수, class 의 수)
```

```python
print('결정 함수의 결과(첫 6개) : \n {}'.format(gbrt.decision_function(X_test)[:6,:]))
# 각 열 : 클래스에 대한 확신의 정도
# 값이 크면 클래스에 속할  확률이 높은 것이다.
# 예측 클래스는 : 가장 확률 값이 큰 것으로 예측한다.
```

```python
print('가장 큰 결정 함수의 인덱스 : \n {}'.format(
        np.argmax(gbrt.decision_function(X_test), axis = 1)))
```

```python
# predict.proba() : 위와 형태는 같고 각 행의 합의 1 이다. 
# 각 클래스에 속할 확률을 표현한다.

print('예측 확률 : \n {} '.format(gbrt.predict_proba(X_test)[:6,:]))
# 행 방향으로 확률을 더한다.
print('합 : {}'.format(gbrt.predict_proba(X_test)[:6,:].sum(axis = 1)))
```

```python
# predict_proba 에 argmax 를 적용하여 예측을 재현할 수 있다.
# 각 데이터 포인트(행)에서 가장 큰 확률값(predict_proba)를 갖는 클래스로 예측하는 것
print('predict_proba 와 argmax 를 사용하여 \n예측한 클래스 : \n {}'.format(np.argmax(gbrt.predict_proba(X_test), axis = 1)))
```

```python
# 제공되는 예측함수를 사용하여 결과와 비교하면
print('예측 클래스 : \n {}'.format(gbrt.predict(X_test)))

# 예측한 값이 같게 된다.
# 즉, 작동원리는 각 행에 대해 가장 큰 확률 값을 갖는 클래스로 예측한다.
```

```python
# 위와 같이 클래스 이름 자체가 정수로 된 경우 문제가 없지만
# 클래스 이름이 문자열이나 다른 형식으로 된 경우  .calsses_ 를 통해 클래스 이름을 알아야한다.
# predict_proba, argmax 를 사용하여 얻은 분류 클래스 : 0,1,2....
# predict를 사용하여 얻은 분류 클래스 : 'setosa'....


# iris 데이터 셋의 타깃을 클래스의 이름으로 나타내기
named_target = iris.target_names[y_train] # ['이름0' , '이름1', '이름2'][0 or 1 or 2 가 요소인 리스트] : 정수를 이름과 매칭한 리스트
named_target
```

```python
logreg = sklearn.linear_model.LogisticRegression()

logreg.fit(X_train, named_target) # target 변수를 클래스 이름으로 준다. 'setosa'....
print('훈련 데이터에 있는 클래스의 종류 : {}'.format(logreg.classes_)) # classes_를 통해 class의 이름을 알아낸다.
print('예측 : {}'.format(logreg.predict(X_test)[:10]))

argmax_dec_func = np.argmax(logreg.decision_function(X_test), axis = 1)
print('가장 큰 결정 함수의 인덱스 : {}'.format(argmax_dec_func[:10]))

#  1 : versicolor
# 0 : setosa
# 2 : virginica
```
