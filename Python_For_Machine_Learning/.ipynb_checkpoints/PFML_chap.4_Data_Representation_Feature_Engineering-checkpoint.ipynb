### 주의 사항
#### 데이터셋 분리 시 주의 사항
#### 처음부터 분리 후 처리를 하게 되면, 두 데이터셋(train, test) 에서 나타나는 범주가 다를 수 있다.
#### -> train 에 없는 범주가 test 에서 나오는 경우 에러를 발생시킨다.
#### 따라서 전체 데이터셋(train + test) 에서 get_dummies 로 만든 후 
#### 그 데이터 셋을 분리해야 범주의 부재로 인한 에러가 발생하지 않는다.

## 4.2 구간 분할, 이산화 & 선형 모델, 트리 모델
### 사용 모델에 따라 데이터 표현방식이 중요 -> 성능에 영향을 미친다.
### 선형 모델, 트리모델

```python
### wave 데이터셋으로 해보자 / input feature 가 1개이다.
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

X, y = mglearn.datasets.make_wave(n_samples = 100 )
line = np.linspace(-3,3, 1000, endpoint = False).reshape(-1,1) # reshape 에서 -1 부분은 몇 차원인지 모를 때 쓰는 것 : 옆의 차원에 맞춰서 유연하게 바뀐다.

reg = LinearRegression().fit(X,y)
plt.plot(line, reg.predict(line), '--', label = 'liner regression')


reg = DecisionTreeRegressor(min_samples_split = 3).fit(X,y)
plt.plot(line, reg.predict(line), label = 'DecisionTree regressor')

plt.plot(X[:,0],y, 'o', c = 'k')
plt.ylabel('output of regression')
plt.xlabel('input feature')
plt.legend(loc = 'best')

# 입력 특성이 1개인 경우 
# 선형 모델은 직선으로 나타난다.
```

### 이산화(binning) 을 사용하여 
### 하나의 입력 특성 -> 여러 개의 feature 로 만들기
#### np.linsapce(시작 , 끝, 절단점의 수)
#### np.digitize() : 각 데이터 포인트가 어떤 구간에 속하는지
#### OneHotEncoder(scitkit-learn 패키지) : 숫자로 된 범주형 변수를 one-hot-encoder 한다.

```python
bins = np.linspace(-3,3,11) #
print('구간 : {}'.format(bins))

# 1-3 부터 3까지 11개의 절단점을 만들어 10개의 구간을 만든다
# 1번 구간 : -3 ~ -2.4 
```

```python
which_bin = np.digitize(X, bins = bins) # 데이터 포인트가 어느 구간에 속하는지
print('\n 데이터 포인트(5번째 까지) : \n ', X[:5])
print('\n 데이터 포인트의 소속 구간 : \n ', which_bin[:5])

# 연속형 변수 - 어느 구간에 속하는지 0,1 코딩을 통해 범주화
```

```python
# OneHotEncoder 를 통해 one-hot-encoding 을 한다.
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse = False)
encoder.fit(which_bin) # which_bin 에 나타난 유일값을 찾는다. 데이터가 속해있는 구간들을 열로 바꾸기 위한 전 단계

X_binned = encoder.transform(which_bin)
print(X_binned[:5]) # 5번째 데이터 포인트까지 어느 구간에 속하는지 보여준다.
```

```python
# 각 data point 100개에 대해
# 구간을 feature 처럼 사용하여 어느 구간에 속하는지 0, 1 coding으로 나타낸다.
X_binned.shape
```

```python
line_binned = encoder.transform(np.digitize(line, bins = bins))

# 구간화 된 one-hot-encoding 데이터를 입력 data 로 주어 회귀를 실행한다.
reg = LinearRegression().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label = 'binned linear regression')

reg = DecisionTreeRegressor(min_samples_split = 3).fit(X_binned , y)
plt.plot(line, reg.predict(line_binned), '--', label = 'binned DecisionTree reg')

plt.plot(X[:,0], y, 'o', c = 'k')
plt.vlines(bins, -3, 3, linewidth = 1, alpha = .2)
plt.legend(loc = 'best')
plt.ylabel('regression output')
plt.xlabel('input feature')

# 하나의 직선을 만든 선형 모델이 결정트리와 같은 예측을 하였다.
# 각 구간마다 상수값을 예측한다.

# 선형 모델 : 더 유연해짐
# 결정 트리 : 덜 유연해짐
### 결정 트리 자체가 가장 좋은 분류 구간을 찾는 것
### 또한 한번에 여러 특성을 고려할 수 있음
```

## 4.3 상호작용과 다항식
### feature 를 풍부하게 하는 것 

```python
# 앞의 방법은 구간마다 상숫값을 예측하였다.
# 각 구간마다 기울기를 추가하면 더 유연해질 것(원본 데이터 열(X)를 추가하여 기울기를 추가해준다.)
# 
X_combined = np.hstack([X, X_binned]) # hstack() : 내부의 row 길이가 같은 요소를 열별로 합친다. (열 추가)
X_combined.shape
```

```python
reg = LinearRegression().fit(X_combined, y)

line_combined = np.hstack([line, line_binned])

plt.plot(line, reg.predict(line_combined), label = 'regression which is combined with original feature')

for bin in bins:
    plt.plot([bin, bin], [-3,3], ':', c = 'k', linewidth = 1)
    
plt.legend(loc = 'best')
plt.ylabel('output of reg')
plt.xlabel('input feature')
plt.plot(X[:,0], y, 'o', c = 'k')

# 각 구간마다 예측값이 상수가 아닌 직선이 되었다. 하지만 모든 구간에 같은 기울기를 가진다.
```

```python
# 곱 특성을 추가하여
X_product = np.hstack([X_binned , X * X_binned])
print(X_product.shape)
```

```python
reg = LinearRegression().fit(X_product, y)

line_product = np.hstack([line_binned,  line * line_binned])

plt.plot(line, reg.predict(line_product), label = 'regression which is producted with original feature')

for bin in bins:
    plt.plot([bin, bin], [-3,3], ':', c = 'k', linewidth = 1)
    
plt.legend(loc = 'best')
plt.ylabel('output of reg')
plt.xlabel('input feature')
plt.plot(X[:,0], y, 'o', c = 'k')

# 구간마다 다른 기울기로 예측한다.
```

### 적은 특성에서 여러 특성을 만들어 유연하게 하는 법
### 1. 구간 나누기
### 2. 회귀 변수에 원래 특성의 몇 제곱의 열을 추가 하기
### 3. 특성끼리의 곱에 대한 것을 특성으로 추가하기
### PolynomialFeatures(degree = ,  include_bias = ) 를 사용하여 만들 수 있다.
#### degree : 최고차항의 수 (원본 데이터의 feature 들의 조합 중 가장 크게 만들 수 있는 항의 차수)
#### 새로 만들어진 열의 조합을 보고 싶은 경우 인스턴스명.get_feature_names 로 호출

```python
# X ** 10 까지의 고차항을 만든다.

poly = sklearn.preprocessing.PolynomialFeatures( degree = 10, include_bias =False) # include_bias = True : 절편을 위해 값이 1 인 열을 추가
poly.fit(X) # X에 맞는 프로세싱 방법 익힘

X_poly = poly.transform(X) # 위의 익힌 방법으로 X 데이터를 변환
```

```python
print(X_poly.shape)
```

```python
print('X의 원소 : \.n {}'.format(X[:5,:]))
print('X_poly 의 원소 : \n {}'.format(X_poly[:5,:])) # X_poly 의 각 열은 X**n 을 나타낸다.
```

```python
# 차수를 알려주는 get_feature_names
# plynomialfeatures 객체의 하위 메서드

poly.get_feature_names()
```

```python
# 이를 이용한 다항회귀 모델
reg = LinearRegression().fit(X_poly, y)

line_poly = poly.transform(line)
plt.plot(line, reg.predict(line_poly), label = 'multilinear regression')
plt.plot(X[:,0], y, 'o', c= 'k')
plt.ylabel('ouput of regression')
plt.xlabel('input features')
plt.legend(loc = 'best')

# 부드러운 곡선
# 하지만 차수가 너무 높고, 특성이 많다보니 너무 민감하게 반응 ( 일반화 부족)
```

```python
# 비교 대상 : 원본 데이터에 대한 커널 SVM 모델
from sklearn.svm import SVR

for gamma in [1,10]:
    svr = SVR(gamma = gamma).fit(X,y)
    plt.plot(line, svr.predict(line), label = 'SVR gamma = {}'.format(gamma))
    
plt.plot(X[:,0], y, 'o', c = 'k')
plt.ylabel('regression ouput')
plt.xlabel('input features')
plt.legend(loc = 'best')

# 매개변수 gamma 가 클수록 더 민감하게 반응
```

## 4.4 일변량 비선형 변환
### 다항회귀 말고 수학함수(sin, cos, exp...) 를 이용한 feature 추가, 변환
### 트리 모델 : feature 의 순서에 영향 받음
### 선형 모델, 신경망 모델 : 스케일, 분포에 영향 받음 / 이때 log, exp 로 스케일 변환을 많이 해준다.
### 대부분의 모델 : 각 feature 가 정규분포일 때 성능이 좋음
#### 따라서 정상성을 따르지 않는 경우 exp, log 를 통해 강제로 변환해준다.


```python
rnd = np.random.RandomState(0)

X_org = rnd.normal(size = (1000, 3))
w = rnd.normal(size = 3)

X = rnd.poisson(10 * np.exp(X_org)) # X 는 포아송분포
y = np.dot(X_org, w)

print(X[:10, 0]) # 첫번째 특성의 앞 10개의 값
```

```python
# X 의 각 값의 빈도를 세자

print('특성 출현 횟수 : \n {}'.format(np.bincount(X[:,0]))) # 첫번째 특성에 등장하는 값들의 빈도 수
```

```python
bins = np.bincount(X[:,0])
plt.bar(range(len(bins)), bins, color = 'grey')
plt.ylabel('frequency')
plt.xlabel('value')
# 지수분포 형태
# 정규 분포 형태와 거리가 멀다 : 선형회귀에서는 이러한 데이터를 잘 처리하지 못함
```

```python
# 정규 분포 형태와 거리가 멀다 : 선형회귀에서는 이러한 데이터를 잘 처리하지 못함
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(X, y , random_state = 0)

score = Ridge().fit(X_train,y_train).score(X_test, y_test)

print('테스트 점수 : {:.2f}'.format(score))

# 좋지 않은 결과..
```

```python
#입력 값을 log 변환을 하면 정상성에 가까워진다.
X_train_log = np.log(X_train + 1)
X_test_log = np.log(X_test + 1)

plt.hist(X_train_log[:,0], bins = 50, color = 'gray')
```

```python
# 입력 변수의 분포 형태를 조정한 후(정규분포에 가깝도록) 모델을 fitting 하면 훨씬 좋은 결과가 나온다.

score = Ridge().fit(X_train_log, y_train).score(X_test_log, y_test)
print('log 변환으로 스케일 조정 후 score : {:.2f}'.format(score))


#0.66 -> 0.87 로 성능이 증가하였다
```

## 4.5.1 일변량 통계
### ANOVA (분산 분석) 임 
### target 과 연관이 있는 변수(feature)를 찾는거지 / 각 feature 가 독립적으로 평가된다.
### 따라서 입력(X) 와 출력(y) 를 인자로 주어야 영향 있는 feature 를 찾는다.
### feature선택 시 선택 기준
#### 분류 : f_classif(default)
#### 회귀 : f_regression
### feature 개수 선택
#### SelectKBest : k 개를 정해서 선택 가능
#### SelectPercentile : 원본데이터에서 원하는 비율만큼 선택 가능
### 선택된 특성 확인(영향 있는 특성)
#### 인스턴스명.get_support()


```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectPercentile
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

# 고정된 난수 발생
rng = np.random.RandomState(42)
noise = rng.normal(size = (len(cancer.data), 50))
# 데이터 노이즈 특성을 추가
# 처음에 30객는 원본특성, 다음 50개는 노이즈
X_w_noise = np.hstack([cancer.data, noise]) # 특성이 80개인 데이터(30 + 50)


X_train, X_test, y_train, y_test = train_test_split(X_w_noise, cancer.target, random_state = 0 ,test_size = .5)

# f_classif(기본값) 와 SelectPercentile 을 사용하여 특성의 50% 선택

select = SelectPercentile(percentile = 50) # 선택비율을 설정해주며 인스턴스 생성
select.fit(X_train, y_train) # 학습 데이터에 대해 중요한 feature 를 찾는다.

X_train_selected = select.transform(X_train) # 학습 input 에 대해 적용하여 feature 를 추출한다.

print('X_train 형태 : {}'.format(X_train.shape))
print('X_train_selected 형태 : {}'.format(X_train_selected.shape))

# feature 가 80개애서 40개(50%)로 줄어들었다.
```

```python
# 인스턴스명.get_support() 를 통해 선택된 feature를 알 수 있다.
mask = select.get_support()
mask

plt.matshow(mask.reshape(1,-1), cmap = 'gray_r')
plt.xlabel('feature number')

# 주로 원본데이터의 feature 가 선택된 것을 볼 수 있다.
```

```python
# 전체특성(80개) vs 선택특성(40개) 에서의 성능비교
# 로지스틱 회귀 사용
X_test_selected = select.transform(X_test)

from sklearn.linear_model import LogisticRegression


for train_input, score_input,  string in zip([X_train, X_train_selected], [X_test, X_test_selected], ['전체 특성' , '선택 특성']):
    score = LogisticRegression().fit(train_input, y_train).score(score_input, y_test)
    print('{} 에서의 점수 : {:.2f}'.format(string, score))
     
# 차이가 크지는 않지만(노이즈 때문) 선택 특성에서 더 좋은 점수가 나온다.
```

## 4.5.2 모델 기반 특성 선택
### 지도 학습을 통한 중요한 feature 선택
### 특성의 중요도 순서를 매길 수 있어야 한다.
#### feature_importance_ 속성
### 트리모델의 상위노드, 선형 모델의 계수의 절댓값... 선형 모델에서 L1 규제 적용 후 살아남는 것
### 일변량과 차이점 : feature 간 상호작용을 잡아낼 수 있다.
### SelectFromModel(지도학습 모델 인스턴스 , threshold = '임계치')
#### 임계치를 넘는 feature 들을 반환한다.

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

select = SelectFromModel(
        RandomForestClassifier(n_estimators = 100, random_state = 42),
        threshold = 'median')
```

```python
select.fit(X_train, y_train) # 비지도 학습으로 target 에 대한 중요 feature 를 찾는 .과정
X_train_l1 = select.transform(X_train)
print('X_train 형태 : {}'.format(X_train.shape))
print('X_train_l1 형태 : {}'.format(X_train_l1.shape))

# feature 의 수가 80 -> 40개
```

```python
# 선택된 특성 그려보기
mask = select.get_support()
plt.matshow(mask.reshape(1,-1), cmap = 'gray_r')
plt.xlabel('feature number')

# 주로 원본 데이터의 feature 가 선택되었다.
# 원본 데이터(feature number 29까지) 에서 2개를 제외하고 선택됌
```

```python
# 성능 테스트
X_test_l1 = select.transform(X_test)

print('전체 feature 에서의 점수 : {:.2f}'.format(LogisticRegression().fit(X_train, y_train).score(X_test, y_test))) # 전체 feature 의 점수
print('선택 된 feature 에서의 점수 : {:.2f}'.format(LogisticRegression().fit(X_train_l1, y_train).score(X_test_l1, y_test))) # 선택 된 feature 의 점수
```

## 4.5.3 반복적 특성 선택
### 특성 수가 다른 일련의 모델을 만든다.
### 1. 전진 선택법
### 2. 후진 소거법
#### 단계마다(feature 가 다른 모델) 계산해야 하기 때문에 계산 비용이 크다.
#### RFE(모델 인스턴스, n_feature_to_select = 선택할 feature 수)

```python
from sklearn.feature_selection import RFE

select = RFE(RandomForestClassifier(n_estimators = 100, random_state = 42),
                n_features_to_select = 40)

select.fit(X_train, y_train)
mask = select.get_support()

plt.matshow(mask.reshape(1,-1), cmap = 'gray_r')
plt.xlabel('feature numbers')

# 오래걸린다. 80 feature 에서 40개로 줄여야 하므로 
# randomforest 모델로 총 40번 학습한다.
# 원본 feature (29까지) 중에 1개만 놓쳤다.
```

```python
X_train_rfe = select.transform(X_train)
X_test_rfe = select.transform(X_test)


for input_train, input_test, string in zip([X_train, X_train_rfe], [X_test, X_test_rfe],['전체 특성', 'RFE 로 골라낸 특성']):
    score = LogisticRegression().fit(input_train, y_train).score(input_test , y_test)
    print('{} 모델 점수 : {:.2f}'.format(string, score))
    
```

## 4.6 전문가 지식 활용
### 전문가의 지식 : 초기 데이터에서 유용한 feature 를 골라낼 때 좋다.
#### 해당 도메인의 중요한 요소를 잡아줄 수 있기 때문

```python
# 예제 : 앤디의 자전거 대여
citibike = mglearn.datasets.load_citibike()
```

```python
citibike.head()
```

```python
# 8월 한달동안의 대여 횟수 그래프
plt.figure(figsize = (10,3))
xticks = pd.date_range(start = citibike.index.min(), end = citibike.index.max(), 
                               freq = 'D') # 하루를 주기로 datetime 객체를 만든다.

week = 'Sun,Mon,Tue,Wed,Thur,Fri,Sat'.split(',')
xticks_name = [week[int(w)] + d for w, d in zip(xticks.strftime('%w'), xticks.strftime(' %m-%d'))]
# strftime('format형식') : format 형식에 맞는 문자열을 반환한다.
# strftime('%ww') : 앞의 datetime 에 해당하는 요일의 인덱스를 뽑아준다.(0 - 일요일 , 1- 월요일...)
plt.xticks(xticks, xticks_name, rotation = 90, ha = 'left')
plt.plot(citibike, linewidth = 1)
plt.xlabel('date')
plt.ylabel('rent frequency')
```

```python
# 시계열 예측을 해보자
# 입력 : 시간(날짜)
# 출력 : 대여 횟수(예측)

# 날짜를 저장하는 타입 : POSIX( 1970/1/1 00:00:00 부터 지금까지 경과한 sec)
```

```python
# 데이터 준비
# y
y = citibike.values

# X : POSIX 시간을 10**9 로 나누어 반환
X = citibike.index.astype('int64').values.reshape(-1,1) // 10**9
```

```python
# 시계열 데이터는 train_test_split 처럼 랜덤하게 분리하면 안된다.
# 시간에 영향을 받으므로 순서에 맞게 분리 해야한다.
# 앞의 184개 : train
# 뒤의 것 : test

n_train = 184

def eval_on_feature(features ,target, regressor):
    # 훈련 , 테스트 분리
    X_train, X_test = features[:n_train], features[n_train:]
    y_train, y_test = target[:n_train], target[n_train:]
    
    regressor.fit(X_train, y_train)
    print('테스트 세트 R^2 : {:.2f}'.format(regressor.score(X_test, y_test)))
    y_pred = regressor.predict(X_test) # test 세트로 예측
    y_pred_train = regressor.predict(X_train) # train 세트로 예측 : 과적합 판단하기
    
    
    plt.figure(figsize = (10,3))
    
    plt.xticks(range(0,len(X), 8), xticks_name, rotation = 90, ha = 'left')
    
    # 각 그래프를 그린다.
    plt.plot(range(n_train), y_train, label = 'train') #  train 데이터의 target 값(0~184)
    plt.plot(range(n_train, len(y_test) + n_train), y_test, '-', label = 'test' ) # test 데이터의 target 값(184~)
    plt.plot(range(n_train), y_pred_train, '--', label = 'prediction(train)')
    plt.plot(range(n_train, len(y_test) + n_train), y_pred, '--', label = 'prediction(test)')
    
    plt.legend(loc = (1.01, 0))
    plt.xlabel('date')
    plt.ylabel('rate frequency')
```

```python
from sklearn.ensemble import RandomForestClassifier

regressor = RandomForestClassifier(n_estimators = 100, random_state = 0)
eval_on_feature(X,y,regressor)

# 성능이 매우 별로이다 
## 이유 : 외삽의 능력이 없다(POSIX 로 볼 때 train 에 속하지 않는 시간을 input 으로 주어 테스트했기 때문)
## 결국 : test 데이터의 시간대와 가장 가까이 있는 값으로 예측함
```

```python
# 전문가의 시각
## 처음 그래프를 보면, 시간과 요일이 중요하다는 것을 파악할 수 있었다.
X_hour = citibike.index.hour.values.reshape(-1,1)
eval_on_feature(X_hour, y, regressor)

# 성능이 조금 더 좋아졌다.
# 하지만 주간 패턴은 예측 못하듯
```

```python
# 요일 정보도 추가
X_hour_week = np.hstack([citibike.index.dayofweek.values.reshape(-1,1),
                                 citibike.index.hour.values.reshape(-1,1)])

eval_on_feature(X_hour_week, y, regressor)

# 이제 하루의 시간, 요일에 따른 주기적인 패턴
```
